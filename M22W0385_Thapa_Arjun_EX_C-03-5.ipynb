```python
############################################################
# Play with human - Tic-Tac-Toe Game Controller
# (c) 2022, Masaharu Imai
# Version 1.0
# Last modification: 2022-12-07
############################################################

from Tic_Tac_Toe import *

def game_control(board, start_pos, num_games, player1, player2):
    win_count = 0
    for i in range(num_games):
        board[start_pos] = 1
        (result, _) = play_game(board, player1, player2)
        if result == 1:
            win_count += 1
        board[start_pos] = 0 
    return win_count / num_games


def ask_place(board):
    while True:
        ans = input("Where to place your piece? ")
        pos = int(ans)
        if not(0 <= pos <= 8):
            print("Sorry,", pos, "is out of range.") 
            print("\n0 1 2\n3 4 5\n6 7 8\n")
            print("Enter 0 - 8")
        elif board[pos] != 0:
            print("Sorry,", pos, "is already occupied.") 
            print("\n0 1 2\n3 4 5\n6 7 8\n")
            print("Enter 0 - 8")
        else:
            return pos

def human_player(board, player):
    print("Your turn.")
    print()
    show_board(board)
    print()
    return ask_place(board)    

############################################################
# Testbench for Play_with_human
############################################################

if __name__ == "__main__":
    
    board = [0] * 9
    print("Let's play Tic-Tac-Toe!")
    print()
    ans  = input("Do you want to play first? (y/n) ")
    if ans in ["y", "Y"]:
        human = 1
        pos = ask_place(board)
        board[pos] = 1
        player_1 = human_player
        player_2 = player_Random
    else:
        human = 2
        player_1 = player_Random
        player_2 = human_player
        
    (result, record) = \
        play_game(board, player_1, player_2)
    
    turn = 3 - human
    for p in record:
        board[p] = turn
        turn = 3 - turn
    print()
    show_board(board)
    print()
    
    if result == 0:
        print("Draw")
    elif result == human:
        print("Congratulations! You won the game.")
    else:
        print("Sorry, you lost the game.")
        
############################################################
# End of Play with human
############################################################
```
