```python
##################################################
# Testbench for candidate()
##################################################

def candidate(board):
    candidates = []
    for i in range(9):
        if board[i] == 0:
            candidates.append(i)
    return candidates


fname  = "testvec_candidate.txt"

def decode(line):
    # line is separated by " "
    # Format: pieces(int*8) /
    #         candidate(int**) 
    data = line.split("/")
    board  = [int(d) for d in data[0].split()]
    cand   = [int(d) for d in data[1].split()]
    return (board, cand)

def show_board(board):
    mark = ("-", "O", "X")
    for r in range(3):
        for c in range(3):
            print(mark[board[r*3+c]], end=" ")
        print()
    
success = True
with open(fname) as fin:
    lines = fin.readlines()
    no = 1
    for line in lines:
        if line[0]=="#": # Skip comment
            continue
        
        (board, exp) = decode(line)
        print(f"{board = }, {exp = }")
        cand = candidate(board)
        print(f"No = {no}, {board = }")
        print(f"{cand = }  ", end="")
        if cand == exp:
            print("OK")
        else:
            success = False
            print("NG")
        print()
        show_board(board)
        print()
        no += 1

if success:
    print("Congratulations!")
else:
    print("Try again!")
```
