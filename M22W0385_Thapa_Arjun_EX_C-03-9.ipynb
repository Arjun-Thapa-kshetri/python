```python
from Tic_Tac_Toe import *

def checkDraw(board):
    for i in range(9):
        if (board[i] == 0):
            return False
    return True

def M22W0385(board, player):
    board = list(board)
    bestScore = -800
    bestMove = 0

    for i in range(9):
        if (board[i] == 0):
            board[i] = player
            score = minimax(board, 0, False)
            board[i] = 0
            if (score > bestScore):
                bestScore = score
                bestMove = i

    return bestMove



def minimax(board, depth, isMaximizing):
    if (is_win(board, 1)):
        return 1
    elif (is_win(board, 2)):
        return -1
    elif (checkDraw(board)):
        return 0

    if (isMaximizing):
        bestScore = -800
        for i in range(9):
            if (board[i] == 0):
                board[i] = 1
                score = minimax(board, depth + 1, False)
                board[i] = 0
                if (score > bestScore):
                    bestScore = score
        return bestScore

    else:
        bestScore = 800
        for i in range(9):
            if (board[i] == 0):
                board[i] = 2
                score = minimax(board, depth + 1, False)
                board[i] = 0
                if (score < bestScore):
                    bestScore = score
        return bestScore


if __name__ == "__main__":
    def player_1(board, player):
        return M22W0385(board, player)


    def player_2(board, player):
        return player_Random(board, player)


    pieces = "000000000"
    game_control(10, player_1, player_2, pieces, Debug=3)
```
