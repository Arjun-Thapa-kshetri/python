```python
##################################################
# Testbench for is_win()
##################################################
fname  = "testvec_is_win.txt"

def is_win(board, player):
    for i in range(3):
        if board[i*3] == player and board[i*3+1] == player and board[i*3+2] == player:
            return True
    for i in range(3):
        if board[i] == player and board[i+3] == player and board[i+6] == player:
            return True
    if board[0] == player and board[4] == player and board[8] == player:
        return True
    if board[2] == player and board[4] == player and board[6] == player:
        return True
    return False

def decode(line):
    # line is separated by " "
    # Format: pieces(int*8)
    #         player(int*1) Win/Lose(char*1) 
    data   = line.split()
    board  = [int(data[i]) for i in range(9)]
    player = int(data[9])
    win    = (data[10]=="W")
    return (board, player, win)

def show_board(board):
    mark = ("-", "O", "X")
    for r in range(3):
        for c in range(3):
            print(mark[board[r*3+c]], end=" ")
        print()
    
success = True
with open(fname) as fin:
    lines = fin.readlines()
    no = 1
    for line in lines:
        if line[0]=="#": # Skip comment
            continue
        
        (board, player, win) = decode(line)
        print(f"No = {no}, {player = }, {win = },")
        print(f"{board = } : ", end="")
        if is_win(board, player) == win:
            print("OK")
        else:
            success = False
            print("NG")
        print()
        show_board(board)
        print()
        no += 1

if success:
    print("Congratulations!")
else:
    print("Try again!")
```
